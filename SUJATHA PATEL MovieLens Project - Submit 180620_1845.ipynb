```python
#since movie id is numerical column, drop it and then sum of the rest of the column group by release year
movie_genre_year = item.drop(columns="movie id").groupby("release year").sum().T
```

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#Merge Data and Item Datasets
data_movie = data.merge(item, how='inner', on='movie id')
```

```python
#Count movies which has more than 100 ratings
data_movie100 = data_movie.groupby('movie id')['rating'].count().reset_index()              
data_movie100_filter = list(data_movie100[data_movie100['rating']>100]['movie id'])
```

```python
#find the average rating, filter only for those in 100 ratings and for top 25
data_rating_avg = data_movie.groupby(['movie id','movie title'])['rating'].mean().sort_values(ascending=False).reset_index()
print("9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100")
data_rating_avg = data_rating_avg[data_rating_avg['movie id'].isin(data_movie100_filter)].head(25)
print(data_rating_avg)
```

```python
plt.figure(figsize=(20,8))

#Seaborn barplot
sns.set(style="darkgrid")
chart_rating = sns.barplot(x='movie title',y='rating',
                          data=data_rating_avg,
                          linewidth=1,
                          palette="Set3",
                          edgecolor=sns.color_palette("dark", 3),ci=None) 
chart_rating.set_xticklabels(chart_rating.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right')
#Annotating bars
for p in chart_rating.patches:
    chart_rating.annotate(format(p.get_height(), '.2f'), 
                          (p.get_x() + p.get_width() / 2., p.get_height()), 
                          ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

#Adding Labels   
plt.title("TOP 25 MOVIES BY RATING [with atleast 100 ratings]", size=22)
plt.ylabel("Rating", Size=18)
plt.xlabel("Movie Title", Size=18)
plt.show()
```
